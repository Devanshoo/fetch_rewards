### Exploding the 'rewardsReceiptItemList' coloumn data from receipts_data table as it contains lot of relevant information and also a key value (brandcode) which we can later use to apply joins fron brand_data table

```python
receipts_data = receipts_data.explode('rewardsReceiptItemList')
```

```python
receipts_data.head()
```

```python
receipts_data.reset_index(inplace=True)
```

```python
receipts_data = receipts_data.fillna({'rewardsReceiptItemList':'{}'})
```

```python
receipts_data.isna().sum()
```

```python
receipts_data = receipts_data.fillna({'rewardsReceiptItemList':'{}'})
```

```python
receipts_data['rewardsReceiptItemList'] = receipts_data['rewardsReceiptItemList'].apply(lambda x:str(x))
receipts_data['rewardsReceiptItemList'] = receipts_data['rewardsReceiptItemList'].apply(literal_eval)
```

```python
receipts_data.rewardsReceiptItemList
```

```python
type(receipts_data.rewardsReceiptItemList[0])
```

```python
rewardsReceiptItemList = pd.json_normalize(receipts_data['rewardsReceiptItemList'],errors='ignore',record_prefix='ItemList')\
.add_prefix('ItemList_')
```

```python
# pd.set_option('display.max_rows', None, 'display.max_columns', None)
rewardsReceiptItemList
```

```python
receipts_merge = pd.merge(receipts_data, rewardsReceiptItemList, left_index=True, right_index=True, how='outer')
```

```python
receipts_merge
```

Now we still have coloumns other than the new exploded from rewardsReceiptItemList which are present in the form of dictionery, hence extracting the key values from their to make the dataset uniform

```python
#Before extracting keys, dropping the NA values from date coloumns as those represent incomplete receipts

receipts_merge = receipts_merge.fillna({"finishedDate":"{}"})
receipts_merge = receipts_merge.fillna({"pointsAwardedDate":"{}"})
receipts_merge = receipts_merge.fillna({"purchaseDate":"{}"})
```

```python
receipts_merge
```

```python
type(receipts_merge['finishedDate'][3])
```

```python
#Going order wise:

receipts_merge['_id'] = receipts_merge['_id'].apply(lambda x: x['$oid'])

receipts_merge['createDate'] = receipts_merge['createDate'].apply(lambda x: x['$date'])
# receipts_merge['createDate'] = receipts_merge['createDate'].apply(literal_eval)

receipts_merge['dateScanned'] = receipts_merge['dateScanned'].apply(lambda x: x['$date'])
# receipts_merge['dateScanned'] = receipts_merge['dateScanned'].apply(literal_eval)

receipts_merge['finishedDate'] = receipts_merge['finishedDate'].apply(lambda x: 0 if type(x) == str else x['$date'])
# receipts_merge['finishedDate'] = receipts_merge['finishedDate'].apply(literal_eval)

receipts_merge['modifyDate'] = receipts_merge['modifyDate'].apply(lambda x: x['$date'])
# receipts_merge['modifyDate'] = receipts_merge['modifyDate'].apply(literal_eval)

receipts_merge['pointsAwardedDate'] = receipts_merge['pointsAwardedDate'].apply(lambda x: 0 if type(x) == str else x['$date'])
# receipts_merge['pointsAwardedDate'] = receipts_merge['pointsAwardedDate'].apply(literal_eval)

receipts_merge['purchaseDate'] = receipts_merge['purchaseDate'].apply(lambda x: 0 if type(x) == str else x['$date'])
# receipts_merge['purchaseDate'] = receipts_merge['purchaseDate'].apply(literal_eval)
```

```python
receipts_merge
```

```python
type(receipts_merge['createDate'][0])
```

```python
import datetime 
receipts_merge['createDate'] = receipts_merge['createDate'].apply(lambda x:datetime.datetime.fromtimestamp(x/1000))
receipts_merge['dateScanned'] = receipts_merge['dateScanned'].apply(lambda x:datetime.datetime.fromtimestamp(x/1000))
receipts_merge['finishedDate'] = receipts_merge['finishedDate'].apply(lambda x:datetime.datetime.fromtimestamp(x/1000))
receipts_merge['modifyDate'] = receipts_merge['modifyDate'].apply(lambda x:datetime.datetime.fromtimestamp(x/1000))
receipts_merge['purchaseDate'] = receipts_merge['purchaseDate'].apply(lambda x:datetime.datetime.fromtimestamp(x/1000))
receipts_merge['pointsAwardedDate'] = receipts_merge['pointsAwardedDate'].apply(lambda x:datetime.datetime.fromtimestamp(x/1000))
```

```python
receipts_merge
```

```python
#Dropping the dict field rewardsReceiptItemList which is present now in the exploded format

receipts_merge = receipts_merge.drop(columns= 'rewardsReceiptItemList')
```

```python
#Converting coloumn names to lowercase

receipts_merge.columns = receipts_merge.columns.str.lower()
```
