Exploding the rewardsReceiptItemList coloumn from receipts_data table as it contains lot of relevant information and also a key value (brandcode) which we can later use to apply joins fron brand_data table

```python
receipts_data = receipts_data.explode('rewardsReceiptItemList')
```

```python
receipts_data
```

```python
receipts_data.reset_index(inplace=True)
```

```python
receipts_data = receipts_data.fillna({'rewardsReceiptItemList':'{}'})
receipts_data['rewardsReceiptItemList'] = receipts_data['rewardsReceiptItemList'].apply(lambda x:str(x))
receipts_data['rewardsReceiptItemList'] = receipts_data['rewardsReceiptItemList'].apply(literal_eval)
```

```python
receipts_data.rewardsReceiptItemList
```

```python
type(receipts_data.rewardsReceiptItemList[0])
```

```python
rewardsReceiptItemList = pd.json_normalize(receipts_data['rewardsReceiptItemList'],errors='ignore',record_prefix='ItemList')\
.add_prefix('ItemList.')
```

```python
# pd.set_option('display.max_rows', None, 'display.max_columns', None)
rewardsReceiptItemList
```

```python
receipts_merge = pd.merge(receipts_data, rewardsReceiptItemList, left_index=True, right_index=True, how='outer')
```

```python
pd.set_option('display.max_columns', None)
receipts_merge
```

Now we still that coloumns other than exploded from rewardsReceiptItemList are present in the form of dictionery and respective keys, extracting just the keys from their to make the dataset uniform

```python
#Before extracting keys, dropping the NA values from date coloumns as those represent incomplete receipts

receipts_merge = receipts_merge.dropna(subset=['finishedDate', 'pointsAwardedDate', 'purchaseDate', 'modifyDate'])
```

```python
#Going order wise:

receipts_merge['_id'] = receipts_merge['_id'].apply(lambda x: x['$oid'])
receipts_merge['createDate'] = receipts_merge['createDate'].apply(lambda x: x['$date'])
receipts_merge['dateScanned'] = receipts_merge['dateScanned'].apply(lambda x: x['$date'])
receipts_merge['finishedDate'] = receipts_merge['finishedDate'].apply(lambda x: x['$date'])
receipts_merge['modifyDate'] = receipts_merge['modifyDate'].apply(lambda x: x['$date'])
receipts_merge['pointsAwardedDate'] = receipts_merge['pointsAwardedDate'].apply(lambda x: x['$date'])
receipts_merge['purchaseDate'] = receipts_merge['purchaseDate'].apply(lambda x: x['$date'])
```
